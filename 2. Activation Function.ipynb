## 2. Activation Function
Let's make a function that computes activation function

### sigmoid

```python
import math
import matplotlib.pylab as plt
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))#1/(1+math.exp(-x))

# output: 0~1
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-56) # very close to 0 
```

```python
x = np.arange(-4, 4, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.ylim(-0.1,1.1)
plt.show()
```

### tanh

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    #(math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

# output: -1~1
```

```python
tanh(-50)
```

```python
tanh(1)
```

```python
x = np.arange(-4, 4, 0.1)
y = tanh(x)

plt.plot(x, y)
plt.ylim(-1.1,1.1)
plt.show()

# when not used numpy in function, it results in 
#'only size-1 arrays can be converted to Python scalars' error
```

### ReLU

```python
def relu(x):
    return np.maximum(0, x) # max(0,x)

# 0 for negative values
# very simple, lightweight -> if not sure which function to use, use ReLU as a default choice!
```

```python
relu(-100)
```

```python
relu(8)
```

```python
x = np.arange(-4.0, 4.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.ylim(-0.1,4)
plt.show()
```

### Leaky ReLU

```python
def leaky_relu(x):
    return np.maximum(0.1*x, x)#max(0.1*x, x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```

```python
x = np.arange(-4.0, 4.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.ylim(-0.5,4)
plt.show()
```
