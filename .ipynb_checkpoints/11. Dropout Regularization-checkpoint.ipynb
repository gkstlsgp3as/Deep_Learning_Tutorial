## 11. Dropout Regularization
When neural network is deeper and broader, it can accompany overfitting. Therefore, we can implement dropout regularization, which is to drop some neurons in the hidden layers.   

### The advantages of this technique:  
#### 1) voting effect:  
The network usually overfits to the mini batch, but it can randomly average the weights, as referred as voting effect, and can apply regularization.  

#### 2) prevent co-adaptation:  
If some specific neurons have high weights or bias, it can slow down other neurons' training speed. Therefore, when we drop out neurons, those specific neurons have strained influence, which can prevent the co-adaptation of neurons. This result in the robust network that does not highly affected by certain dataset.

### Other methods to prevent overfitting
#### 1) data augmentation
When the size of a dataset increases, the model tends to train the general pattern instead of specific pattern or noise.  
Therefore, we can manipulate or add the data, which is referred as a data augmentation. For images, as an example, we can add some noises to the images, rotate the images, or manipute the images. 

#### 2) minimizing the model complexity
The complexity of the neural network is affected by the number of layers, or parameters. Also, we call the number of parameters a capacity. 

#### 3) weight regularization 
-L1 regularization: add the absolute summation of weights to the cost function
-L2 regularization: add the squared summation of weights to the cost function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('sonar_dataset.csv', header=None)
df.sample(5)
```

```python
df.shape
```

```python
df.isna().sum() # they all have values
```

```python
df.columns
```

```python
df[60].value_counts() # binary classfication
```

```python
X = df.drop(60, axis='columns')
y = df[60]
X
```

```python
y
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R -> 1, M -> 0
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
    
])

model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)

## it looks too good!
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10] # this is the answer!
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
## dropout regularization
model = keras.Sequential([
    
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)

## training accuracy has lowered
## but since the dropout is randomly implemented, 
## we cannot guarantee to have the same results
```

```python
model.evaluate(X_test, y_test) # little improvement with the test set
```

```python
y_pred = model.predict(X_test).reshape(-1)
```

```python
y_pred
```

```python
y_pred = np.round(y_pred)
```

```python
y_pred # this is needed to compare directly with the true y
```

```python
print(classification_report(y_test, y_pred))
```

### Results
We could see that dropout regularization prevented the overfitting toward the training set, and improved the accuracy toward the test set. 

And the common value for dropout is 0.5 for hidden layers and a value closed to 1.0 for layers closer to the visible layer. Therefore, I also tried with different dropout value as below.
