## 5. Gradient Descent
Let's explore gradient descent!

This finds the global minima and update the weight and bias as: 
w1 = w1 - learning rate x d(loss)/d(w1) in each epoch

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

### Split Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age','affordability']],df.insurance,test_size=0.2, random_state=25)
```

```python
## scaling data
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
X_train_scaled # age became 0 ~ 1
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])
# 1: whether a person buy an insurance or not (0 or 1; just one value)
# (2, ): two parameters 

model.compile(optimizer='adam',
              loss='binary_crossentropy', # log loss
              metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=5000)
```

```python
# model evaluation
model.evaluate(X_test_scaled,y_test) # model accuracy is great! almost 100%
```

```python
# predict with test set
model.predict(X_test_scaled) # 1 0 0 0 1 1
```

```python
def sigmoid(x):
    import math
    return 1/(1+math.exp(-x))
```

```python
def predict_func(age, affordability):
    weighted_sum = coef[0]*age + coef[1]*affordability + intercept # w1*age + w2*affordability + bias = weighted sum
    return sigmoid(weighted_sum) # sigmoid was the hidden activation function
```

```python
predict_func(0.47, 1) # same as above! 
```

```python
def log_loss(y_true, y_predict):
    epsilon = 1e-15
    y_predict_ref = [max(i, epsilon) for i in y_predict]
    y_predict_ref = [min(i, 1-epsilon) for i in y_predict_ref]
    y_predict_ref = np.array(y_predict_ref)
    return -np.mean(y_true*np.log(y_predict_ref)+(1-y_true)*np.log(1-y_predict_ref))
```

```python
def sigmoid_numpy(x):
    return 1/(1+np.exp(-x)) # you can apply array for this

sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(age, affordability, y_true, epochs):
    # w1, w2, bias
    # initialization
    w1 = w2 = 1 
    bias = 0
    rate=0.5
    n = len(age)
    
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordability + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        # y_predicted = predict_func(age, affordability)
        
        loss = log_loss(y_true, y_predicted)
        
        w1_d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true)) # xi(y_hat - yi)     
        w2_d = (1/n)*np.dot(np.transpose(affordability), (y_predicted-y_true))
        
        bias_d = np.mean(y_predicted-y_true)
        
        w1 = w1 - rate*w1_d
        w2 = w2 - rate*w2_d
        bias = bias - rate*bias_d
        
        print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
              
    return w1, w2, bias
    
```

```python
def gradient_descent(age, affordability, y_true, epochs, loss_threshold):
    # w1, w2, bias
    # initialization
    w1 = w2 = 1 
    bias = 0
    rate=0.5
    n = len(age)
    
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordability + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        # y_predicted = predict_func(age, affordability)
        
        loss = log_loss(y_true, y_predicted)
        
        w1_d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true)) # xi(y_hat - yi)     
        w2_d = (1/n)*np.dot(np.transpose(affordability), (y_predicted-y_true))
        
        bias_d = np.mean(y_predicted-y_true)
        
        w1 = w1 - rate*w1_d
        w2 = w2 - rate*w2_d
        bias = bias - rate*bias_d
        
        print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        
        if loss<=loss_threshold:
            break
              
    return w1, w2, bias
    
```

### Custom Model 
Building a customized model that works same as before

```python
class myNN:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0

    def fit(self, X, y, epochs, loss_threshold):
        self.w1, self.w2, self.bias = self.gradient_descent(X['age'], X['affordability'], y, epochs, loss_threshold)
    
    def gradient_descent(self, age, affordability, y_true, epochs, loss_threshold):
        # w1, w2, bias
        # initialization
        w1 = w2 = 1 
        bias = 0
        rate=0.5
        n = len(age)

        for i in range(epochs):
            weighted_sum = w1 * age + w2 * affordability + bias
            y_predicted = sigmoid_numpy(weighted_sum)
            # y_predicted = predict_func(age, affordability)

            loss = log_loss(y_true, y_predicted)

            w1_d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true)) # xi(y_hat - yi)     
            w2_d = (1/n)*np.dot(np.transpose(affordability), (y_predicted-y_true))

            bias_d = np.mean(y_predicted-y_true)

            w1 = w1 - rate*w1_d
            w2 = w2 - rate*w2_d
            bias = bias - rate*bias_d
            
            if i%50 ==0: # print every 50 iterations
                print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

            if loss<=loss_threshold:
                print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
                # want to know when it breaks
                break

        return w1, w2, bias
    
    def predict(self, X_test):
        weighted_sum = self.w1*X_test['age'] + self.w2*X_test['affordability'] + self.bias
        return sigmoid_numpy(weighted_sum)
    
```

```python
customModel = myNN()
customModel.fit(X_train_scaled, y_train, epochs=500, loss_threshold = 0.4631)
```
